# Data collection (1)


In order to map the mandates of the UN entities, the first step is to create a list of websites of the UN entities based on the United Nations System Chart.From the UN organization chart: https://www.un.org/en/delegate/page/un-system-chart, we noticed that there are 4 major data sources: General Assembly, Security Council, Economic and social council, and Secretariat 


```python
 from requests_html import HTMLSession
session = HTMLSession()

# The unsystem website have informations about most of the UN entities
url = 'http://www.unsystem.org/directory' 

r= session.get(url)
print(r.html.text)
```

```python
r.html.links
```

```python
r.html.absolute_links
```

```python
def get_text_link_from_sel(sel):
    mylist = []
    try:
        results = r.html.find(sel)
        for result in results:
            mytext = result.text
            mylink = list(result.absolute_links)[0]
            mylist.append((mytext, mylink))
        return mylist
    except:
        return None
    print(get_text_link_from_sel(sel))
```

```python
sel = '#main > div > div.panel-panel.panel-col-middle > div > div > div > div > div.view-content > div > ul > li > div > a'
print(get_text_link_from_sel(sel))
```

```python
import pandas as pd
df = pd.DataFrame(get_text_link_from_sel(sel))
df
```

```python
df.to_csv('output.csv', encoding='gbk', index=False) 

#save df to csv file
```

```python
# Fourth - ECOSOC 

# The list of entities related to ECOSOC is listed in the pdf instead of on a website like the above

# Hence, we have to scrape the links and text in the pdf 


import PyPDF2
PDFFile = open("/Users/suetko/Downloads/ecosoc_chart.pdf",'rb')

PDF = PyPDF2.PdfFileReader(PDFFile)
pages = PDF.getNumPages()
key = '/Annots'
uri = '/URI'
ank = '/A'

for page in range(pages):
    print("Current Page: {}".format(page))
    pageSliced = PDF.getPage(page)
    pageObject = pageSliced.getObject()
    if key in pageObject.keys():
        ann = pageObject[key]
        for a in ann:
            u = a.getObject()
            if uri in u[ank].keys():
                print(u[ank][uri])
```
